```python
from sklearn.preprocessing import MinMaxScaler
numeric_columns = data.select_dtypes(include=['int64', 'float64'])
scaler = MinMaxScaler()
data[numeric_columns.columns] = scaler.fit_transform(data[numeric_columns.columns])
print("Normalized Dataset:")
print(data.head())
```

```python
binary_attributes = []
for column in data.columns:
    unique_values = data[column].unique()
    if len(unique_values) == 2 and set(unique_values) == {0, 1}:
        binary_attributes.append(column)
print("Binary Attributes:")
print(binary_attributes)
vector1 = data.iloc[0][binary_attributes].tolist()
vector2 = data.iloc[1][binary_attributes].tolist()
f11 = sum(x & y for x, y in zip(vector1, vector2))
f01 = sum(1 for x, y in zip(vector1, vector2) if x == 0 and y == 1)
f10 = sum(1 for x, y in zip(vector1, vector2) if x == 1 and y == 0)
if f01 + f10 + f11 == 0:
    jc = 0.0
else:
    jc = f11 / (f01 + f10 + f11)
smc_denominator = len(binary_attributes)
if smc_denominator == 0:
    smc = 0.0
else:
    smc = (f11 + sum(1 for x, y in zip(vector1, vector2) if x == y)) / smc_denominator
print("Jaccard Coefficient (JC):", jc)
print("Simple Matching Coefficient (SMC):", smc)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import jaccard_score
from sklearn.metrics import pairwise_distances
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.DataFrame(np.random.randint(0, 2, size=(20, 10)), columns=[f'feature_{i}' for i in range(10)])
def jaccard_similarity(vector1, vector2):
    intersection = np.sum(np.minimum(vector1, vector2))
    union = np.sum(np.maximum(vector1, vector2))
    return intersection / union if union > 0 else 0
num_samples = data.shape[0]
jaccard_matrix = np.zeros((num_samples, num_samples))
smc_matrix = np.zeros((num_samples, num_samples))
cosine_matrix = np.zeros((num_samples, num_samples))
for i in range(num_samples):
    for j in range(num_samples):
        if i == j:
            continue
        vector1 = data.iloc[i].values
        vector2 = data.iloc[j].values
        
        jaccard_matrix[i, j] = jaccard_similarity(vector1, vector2)
        
        smc_matrix[i, j] = (vector1 == vector2).mean()
        
        cosine_matrix[i, j] = 1 - pairwise_distances([vector1, vector2], metric='cosine')[0, 1]

similarity_df = pd.DataFrame({'Jaccard': jaccard_matrix.flatten(),
                              'SMC': smc_matrix.flatten(),
                              'Cosine': cosine_matrix.flatten()})
similarity_df = similarity_df.values.reshape(num_samples, num_samples, -1)
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_df[:, :, 0], annot=True, fmt=".2f", cmap='coolwarm', xticklabels=False, yticklabels=False, cbar_kws={'label': 'Jaccard'}, linewidths=0.5)
plt.title("Jaccard Similarity Heatmap")
plt.show()
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_df[:, :, 1], annot=True, fmt=".2f", cmap='coolwarm', xticklabels=False, yticklabels=False, cbar_kws={'label': 'SMC'}, linewidths=0.5)
plt.title("Simple Matching Coefficient Heatmap")
plt.show()
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_df[:, :, 2], annot=True, fmt=".2f", cmap='coolwarm', xticklabels=False, yticklabels=False, cbar_kws={'label': 'Cosine'}, linewidths=0.5)
plt.title("Cosine Similarity Heatmap")
plt.show()
```
